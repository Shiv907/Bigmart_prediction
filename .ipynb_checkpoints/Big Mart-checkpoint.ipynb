## Feature Engineering

## Item Visibility can't be zero

```python
num_zeros = (data["Item_Visibility"] == 0).sum() / data.shape[0]
num_zeros
```

In our dataset minimum value of Item visibility is zero.Which not make any sense.Every product should be visible  to client.so compute it with mean value.

```python
visibility_item_avg = data.groupby('Item_Identifier')['Item_Visibility'].transform('mean')
data['Item_Visibility'] = data['Item_Visibility'].where(data['Item_Visibility'] != 0, visibility_item_avg)
```

```python
sum(data.Item_Visibility==0)
```

```python
data["Outlet_Years"]=2013-data["Outlet_Establishment_Year"]
data["Outlet_Years"].describe()
```

```python
data['Item_Type'].unique()
```

### Create a broad category of Item Type

```python
data["Item_Type_Combined"]=data["Item_Identifier"].apply(lambda x:x[0:2])
data["Item_Type_Combined"]=data["Item_Type_Combined"].map({"FD":"Food","NC":"Non-Consumable","DR":"Drinks"})
data["Item_Type_Combined"].value_counts()
```

### Change categories of Item_Fat_Content
Replacing values of LF and low fat with Low Fat and reg with Regular

```python
data["Item_Fat_Content"]=data["Item_Fat_Content"].replace({"LF":"Low Fat","low fat":"Low Fat","reg":"Regular"})
data["Item_Fat_Content"].value_counts()
```

#### There are some non consumable that should not be included in Low Fat,fat content so replace them in Non-Edible 

```python
data.loc[data["Item_Type_Combined"]=="Non-Consumable","Item_Fat_Content"]="Non-Edible"
data["Item_Fat_Content"].value_counts()
```

### Feature Transformation

#### The products that are more visible are likely to have higher sales so create new variable Item_Visibility_MeanRatio


```python
# Calculate the mean item visibility for each item identifier
visibility_item_avg = data.groupby("Item_Identifier")["Item_Visibility"].mean()

# Map the mean item visibility values to the original DataFrame
data["Item_Visibility_MeanRatio"] = data["Item_Identifier"].map(visibility_item_avg)

# Calculate the Item_Visibility_MeanRatio by dividing Item_Visibility by the mean visibility
data["Item_Visibility_MeanRatio"] = data["Item_Visibility"] / data["Item_Visibility_MeanRatio"]
```

```python
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)

# Importing dependencies
from feast import FeatureStore
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from joblib import dump


# Getting our FeatureStore
store = FeatureStore(repo_path=".")

# Retrieving the saved dataset and converting it to a DataFrame
training_df = store.get_saved_dataset(name="bigmart_dataset").to_df()
```

```python
def test_train_split(final_df):
    from sklearn.model_selection import train_test_split
    y = final_df['Item_Outlet_Sales']
    X = final_df.drop(labels=['Item_Outlet_Sales', 'event_timestamp', "item_id"], axis=1)
    X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
    return X_train, X_test, y_train, y_test
```

```python
def training_basic_regressor(X_train,y_train):
    from sklearn.ensemble import RandomForestRegressor
    model = RandomForestRegressor()
    model.fit(X_train, y_train)
    return model
```

```python
def predict_on_test_data(model,x_test):
    y_pred = model.predict(x_test)
    return y_pred
```

```python
def get_metrics(y_test,prediction):
    from sklearn import metrics
    R2 = metrics.r2_score(y_test,prediction)
    MAE = metrics.mean_absolute_error(y_test, prediction)
    MSE = metrics.mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(metrics.mean_squared_error(y_test, prediction))
    return {'R2': round(R2, 3), 'MAE': round(MAE, 3), 'MSE': round(MSE, 3), 'RMSE': round(RMSE, 3)}
```

```python
def hyper_parameter_tuning(X_train, y_train):
    # define random parameters grid
    n_estimators = [301,320, 300, 150, 200, 250] # number of trees in the random forest
  #  max_features = ['auto'] # number of features in consideration at every split
    max_depth = [int(x) for x in np.linspace(10, 120, num = 12)] # maximum number of levels allowed in each decision tree
    min_samples_split = [42, 36, 30, 8, 12, 20] # minimum sample number to split a node
    min_samples_leaf = [1, 3, 4, 5, 7,9,11] # minimum sample number that can be stored in a leaf node
    bootstrap = [True, False] # method used to sample data points

    random_grid = {'n_estimators': n_estimators,
                 #   'max_features': max_features,
                    'max_depth': max_depth,
                    'min_samples_split': min_samples_split,
                    'min_samples_leaf': min_samples_leaf,
                    'bootstrap': bootstrap
                  }
    
    from sklearn.model_selection import RandomizedSearchCV
    from sklearn.ensemble import RandomForestRegressor
    regressor = RandomForestRegressor()
    model_tuning = RandomizedSearchCV(estimator = regressor, param_distributions = random_grid,
                   n_iter = 100, cv = 5, verbose=2, random_state=35, n_jobs = -1)
    model_tuning.fit(X_train, y_train)

    print ('Random grid: ', random_grid, '\n')
    # print the best parameters
    print ('Best Parameters: ', model_tuning.best_params_, ' \n')

    best_params = model_tuning.best_params_
    
    n_estimators = best_params['n_estimators']
    min_samples_split = best_params['min_samples_split']
    min_samples_leaf = best_params['min_samples_leaf']
   # max_features = best_params['max_features']
    max_depth = best_params['max_depth']
    bootstrap = best_params['bootstrap']
    
    model_tuned = RandomForestRegressor(n_estimators = n_estimators, min_samples_split = min_samples_split,
                                         min_samples_leaf= min_samples_leaf, #max_features = max_features,
                                         max_depth= max_depth, bootstrap=bootstrap) 
    model_tuned.fit( X_train, y_train)
    return model_tuned,best_params
```

```python
def plot_actual_vs_predicted(y_test, y_pred):
    import matplotlib.pyplot as plt
    plt.figure()
    plt.scatter(y_test, y_pred)
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=2)
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title('Actual vs. Predicted Values')
    plt.savefig('actualVsPred.png')
    plt.close()

def plot_residuals(y_test, y_pred):
    import matplotlib.pyplot as plt
    residuals = y_test - y_pred
    plt.figure()
    plt.scatter(y_pred, residuals)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals')
    plt.title('Residual Plot')
    plt.savefig('residualPlot.png')
    plt.close()

def plot_residuals_distribution(y_test, y_pred):
    import matplotlib.pyplot as plt
    import seaborn as sns
    plt.figure()
    residuals = y_test - y_pred
    sns.histplot(residuals, kde=True)
    plt.xlabel('Residuals')
    plt.ylabel('Frequency')
    plt.title('Distribution of Residuals')
    plt.savefig('residualdistribution.png')
    plt.close()
```

## Create Experiment function for experiment tracking using MLFlow

```python
def create_experiment(experiment_name,run_name, run_metrics,model, actual_vs_predicted_path = None, 
                      residuals_path = None, residual_distri_path = None, run_params=None):
    import mlflow
    import sklearn
    mlflow.set_tracking_uri("http://localhost:5005")
    mlflow.set_experiment(experiment_name)
    
    with mlflow.start_run():
        
        if not run_params == None:
            for param in run_params:
                mlflow.log_param(param, run_params[param])
            
        for metric in run_metrics:
            mlflow.log_metric(metric, run_metrics[metric])
        
        mlflow.sklearn.log_model(model, "RFmodel5", registered_model_name= "RF_Regressor5")
        
        if not actual_vs_predicted_path == None:
            mlflow.log_artifact(actual_vs_predicted_path, 'actualVSpredicted_Plot')
            
        if not residuals_path == None:
            mlflow.log_artifact(residuals_path, "residual_Plot")
            
        if not residuals_path == None:
            mlflow.log_artifact(residual_distri_path, "residual_distribution_Plot")
        
        mlflow.set_tag("tag1", "Random Forest")
        mlflow.set_tags({"tag2":"Hypertuned Model", "tag3":"Production"})
            
    print('Run - %s is logged to Experiment - %s' %(run_name, experiment_name))
```

**Prefer to run it from command line**

`mlflow ui` This will launch mlflow UI in the browser and you can access it using `localhost:5000` but this uses file as backend to store experiments and model artifacts. It does not support model registry functionality. 

To use model registry, you need to have some backend database other than the file system. Ex mysql, sqlite or any other DB mentioned in mlflow docs under backend storage. Refer: https://www.mlflow.org/docs/latest/tracking.html?highlight=scenario#how-runs-and-artifacts-are-recorded

#### In this, we will use sqlite as backend, so Now run this command to start mlflow with backend.
`mlflow server --backend-store-uri sqlite:///mlflow.db --default-artifact-root ./artifacts --host 0.0.0.0 --port 5000`
